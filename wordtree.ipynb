```python
class WTNode:
    def __init__(self,d,l,m,r):
        self.data = d
        self.mult = 0
        self.left = l
        self.right = r
        self.midl = m
          
    # prints the node and all its children in a string
    def __str__(self):  
        st = "("+str(self.data)+", "+str(self.mult)+") -> ["
        if self.left != None:
            st += str(self.left)
        else: st += "None"
        if self.midl != None:
            st += ", "+str(self.midl)
        else: st += ", None"
        if self.right != None:
            st += ", "+str(self.right)
        else: st += ", None"
        return st + "]"
    
    def updateChild(self, oldChild, newChild):
        if self.left == oldChild:
            self.left = newChild
        elif self.right == oldChild:
            self.rigth = newChild
        elif self.midl == oldChild:
            self.midl = newChild
    
class WordTree:
    def __init__(self):
        self.root = None
        self.size = 0
        
    def __str__(self):
        return str(self.root)
    
    def removeRoot(self):
        parentNode = WTNode(None, self.root, None, None)
        self.removeNode(self.root, parentNode)
        self.root = parentNode.left
        self.size -= 1

    # returns the number of times that string st is stored in the tree
    def count(self, st):
        if st == "" or self.root == None:
            return None
        words = [ch for ch in st]
        i = 0
        ptr = self.root
        while i<len(words):
            if ptr == None:
                return 0
            if words[i] == ptr.data:
                if i == len(words)-1:
                    return ptr.mult
                ptr = ptr.midl
                i += 1
            elif words[i] < ptr.data:
                ptr = ptr.left
            else:
                ptr = ptr.right
    
    # returns the lexicographically smallest string in the tree
    # if the tree is empty, return None
    def minst(self):
        if self.root == None:
            return None
        words = []
        ptr = self.root
        while True:
            if ptr.left == None:
                words.append(ptr.data)
                if ptr.mult > 0:
                    break
                else:
                    ptr = ptr.midl
            else:
                ptr = ptr.left
        string = ""
        for i in range(len(words)):
            string += words[i]
        return string
    
    # adds the string st in the tree, increases the size by 1
    # and returns None
    def add(self, st): 
        words = [ch for ch in st]
        if len(words) == 0: return
        i = 0
        if self.root == None:
            self.root = WTNode(words[0], None, None, None)
        ptr = self.root
        while i<len(words):
            if ptr.data == words[i]:
                if i<len(words)-1:
                    if ptr.midl == None:
                        ptr.midl = WTNode(words[i+1], None, None, None)
                    ptr = ptr.midl
                i += 1
            elif words[i] < ptr.data:
                if ptr.left == None:
                    ptr.left = WTNode(words[i], None, None, None)
                ptr = ptr.left
            else:
                if ptr.right == None:
                    ptr.right = WTNode(words[i], None, None, None)
                ptr = ptr.right
        if i == len(words):
            ptr.mult += 1
            self.size += 1
            return
        
    # removes one occurrence of string st from the tree and returns None
    # if st does not occur in the tree then it returns without changing the tree
    # it updates the size of the tree accordingly
    def remove(self,st):
        words = [ch for ch in st]
        i = 0
        if self.root == None or st == "": return
        if len(words)==1 and words[0] == self.root.data and self.root.midl == None:
            if self.root.mult == 1:
                self.removeRoot()
            else:
                self.root.mult -= 1
            return
        parentPtr = None
        ptr = self.root
        while i<len(words)-1 and ptr != None:
            parentPtr = ptr
            if ptr.data == words[i]:
                ptr = ptr.midl
                i += 1
            elif words[i] < ptr.data:
                ptr = ptr.left
            else:
                ptr = ptr.right
        if ptr != None and ptr.midl == None:
            self.size -= 1
            if ptr.mult > 1:
                ptr.mult -= 1
                return
            else:
                self.removeNode(ptr, parentPtr)
        else:
            self.size -= 1
            if ptr.mult > 0:
                ptr.mult -= 1
            return
            
    def removeNode(self, ptr, parentPtr):
        if ptr.right == ptr.left == None:
            parentPtr.updateChild(ptr, None)
        elif ptr.right == None:
            parentPtr.updateChild(ptr, ptr.left)
        elif ptr.left == None:
            parentPtr.updateChild(ptr, ptr.right)
        else:
            parentMinNode = ptr
            minNode = ptr.right
            while minNode.left != None:
                parentMinNode = minNode
                minNode = minNode.left
            if minNode.midl == None:
                parentMinNode.updateChild(minNode, None)
                minNode.left = ptr.left
                minNode.rigth = ptr.rigth
                parentPtr.updateChild(ptr, minNode)
            else:
                newNode = WTNode(minNode.data, ptr.left, None, ptr.right)
                parentPtr.updateChild(ptr, newNode)
        if self.size == 0:
            self.root = None

t = WordTree()
t.add("car"); t.add("can"); t.add("cat"); t.add("cat"); t.add("cat")
print("1: ",t.count("ca"),t.count("can"),t.count("car"),t.count("cat"),t.minst()) 
print(t, t.size)
t.remove("")
print(t, t.size)
t.add("")
print(t, t.size)
t.add("ca")
print("2: ",t.count("ca"),t.count("can"),t.count("car"),t.count("cat"),t.minst()) 
print(t, t.size)
t.remove("car")
print("3: ",t.count("ca"),t.count("can"),t.count("car"),t.count("cat"),t.minst()) 
print(t, t.size)
t.remove("cat"); t.remove("cat"); t.remove("cat")
print("4: ",t.count("ca"),t.count("can"),t.count("car"),t.count("cat"),t.minst()) 
print(t, t.size)

print('')
t.remove("ca"); t.add("car")
print("5: ",t.count("ca"),t.count("can"),t.count("car"),t.count("cat"),t.minst()) 
print(t, t.size)

print("")
t.remove("can"); t.remove("car")
print(t, t.size)
```
